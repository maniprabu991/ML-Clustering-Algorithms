```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3:5].values
```

```python
X
```

```python
from sklearn.cluster import AffinityPropagation
cluster_model = AffinityPropagation(random_state=5).fit(X)
```

```python
cluster_model
```

```python
labels = cluster_model.fit_predict(X)

# Create DataFrame from the synthetic data
supervised = pd.DataFrame(dataset)

# Add cluster labels to DataFrame
supervised['cluster_group'] = labels
```

```python
supervised
```

```python
cluster_centers_indices = cluster_model.cluster_centers_indices_
```

```python
n_clusters = len(cluster_centers_indices)
```

```python
n_clusters
```

```python

plt.figure(figsize=(8, 6))

colors = plt.cycler("color", plt.cm.viridis(np.linspace(0, 1)))
for cluster_index in range(n_clusters):
    cluster_points = X[labels == cluster_index]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {cluster_index}')

for k, col in zip(range(n_clusters), colors):
    class_members = labels == k
    cluster_center = X[cluster_centers_indices[k]]
    for x in X[class_members]:
        plt.plot(
            [cluster_center[0], x[0]], [cluster_center[1], x[1]], color=col["color"]
        )
        
plt.title('Mall Customer Prediction')
plt.xlabel(dataset.columns[3])
plt.ylabel(dataset.columns[4])

plt.show()
```
