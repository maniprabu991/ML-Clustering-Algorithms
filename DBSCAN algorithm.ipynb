```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:,3:5].values
```

```python
X
```

```python
from sklearn.datasets import make_blobs
# Create synthetic data
np.random.seed(7)
x, _ = make_blobs(n_samples=200, centers=1, cluster_std=.3, center_box=(20, 5))
 
# Visualize the data 
plt.scatter(x[:,0], x[:,1])
plt.xlabel(dataset.columns[3])
plt.ylabel(dataset.columns[4])
plt.show()
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps = 0.28, min_samples = 20).fit(X)
print(dbscan.get_params()) 
  
```

```python
# Predict clusters for the data
labels = dbscan.fit_predict(x)
# Visualize the clustered points
plt.scatter(x[:,0], x[:,1], c=labels)
plt.title("DBSCAN Clustering")
plt.xlabel(dataset.columns[3])
plt.ylabel(dataset.columns[4])
plt.show()
```

```python
# Count the number of clusters
n_clusters = len(set(dbscan.labels_))
```

```python
n_clusters
```

```python
labels = dbscan.fit_predict(x)

supervised = pd.DataFrame(dataset)
```

```python
supervised
```

```python
# Assuming supervised DataFrame has fewer samples than the clustering data
# Trim the clustering data to match the length of the DataFrame index
labels_trimmed = labels[:len(supervised)]

# Add cluster labels to DataFrame
supervised['cluster_group'] = labels_trimmed
```

```python
supervised
```

```python
import seaborn as sns

sns.lmplot(
    data=supervised,
    x=supervised.columns[3],
    y=supervised.columns[4],
    hue=supervised.columns[5],
    fit_reg=False,
    facet_kws={'legend_out': True}  # Pass legend_out using facet_kws
)
```
