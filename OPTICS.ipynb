```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3:5].values
```

```python
X
```

```python
from sklearn.cluster import OPTICS
import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)

# Clustering with OPTICS
cluster_model = OPTICS(min_samples=2).fit(X)
```

```python
labels = cluster_model.fit_predict(X)

# Create DataFrame from the synthetic data
supervised = pd.DataFrame(dataset)

# Add cluster labels to DataFrame
supervised['cluster_group'] = labels
```

```python
supervised
```

```python
import seaborn as sns

sns.lmplot(
    data=supervised,
    x=supervised.columns[3],
    y=supervised.columns[4],
    hue=supervised.columns[5],
    fit_reg=False,
    facet_kws={'legend_out': True}  # Pass legend_out using facet_kws
)
```
