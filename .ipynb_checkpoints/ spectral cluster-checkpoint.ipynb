```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3:5].values
```

```python
X
```

```python
from sklearn.cluster import SpectralClustering

# Perform spectral clustering
spectral_clustering = SpectralClustering(n_clusters=2, affinity='nearest_neighbors', assign_labels='discretize',random_state=0).fit(X)
labels = spectral_clustering.fit_predict(X)

# Visualize the clustering result
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.title("Spectral Clustering Result")
plt.xlabel(dataset.columns[3])
plt.ylabel(dataset.columns[4])
plt.show()
```

```python
# Create DataFrame from the synthetic data
supervised = pd.DataFrame(dataset)

# Add cluster labels to DataFrame
supervised['cluster_group'] = labels
```

```python
supervised
```

```python
import seaborn as sns

sns.lmplot(
    data=supervised,
    x=supervised.columns[3],
    y=supervised.columns[4],
    hue=supervised.columns[5],
    fit_reg=False,
    facet_kws={'legend_out': True}  # Pass legend_out using facet_kws
)
```
