```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, 3:5].values
```

```python
X
```

```python
import scipy.cluster.hierarchy as sch

# Ensure your variable X is defined and contains the data you want to cluster
# X = ...

dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Mall Customer Prediction')
plt.xlabel(dataset.columns[3])
plt.ylabel(dataset.columns[4])
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

clusterModel = AgglomerativeClustering(n_clusters=5).fit(X)
```

```python
clusterModel
```

```python
label = clusterModel.fit_predict(X)
```

```python
supervised = pd.DataFrame(dataset)
supervised['cluster_group']=label
supervised
```

```python
import seaborn as sns

sns.lmplot(
    data=supervised,
    x=supervised.columns[3],
    y=supervised.columns[4],
    hue=supervised.columns[5],
    fit_reg=False,
    facet_kws={'legend_out': True}  # Pass legend_out using facet_kws
)
```
