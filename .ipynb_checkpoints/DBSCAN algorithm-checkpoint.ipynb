```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:,3:5].values
```

```python
X
```

```python
from sklearn.cluster import DBSCAN
cluster_model = DBSCAN(eps=3, min_samples=2).fit(X)
```

```python
labels = cluster_model.fit_predict(X)

# Create DataFrame from the synthetic data
supervised = pd.DataFrame(dataset)

# Add cluster labels to DataFrame
supervised['cluster_group'] = labels
```

```python
supervised
```

```python
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler

centers = [[1, 1], [-1, -1], [1, -1]]
X, labels_true = make_blobs(
    n_samples=750, centers=centers, cluster_std=0.4, random_state=0
)

X = StandardScaler().fit_transform(X)
```

```python

n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
print("Total number of cluster:" , n_clusters_)
plt.scatter(X[:, 0], X[:, 1])
plt.title('N_cluster: {}'.format(n_clusters_))
plt.xlabel(dataset.columns[3])
plt.ylabel(dataset.columns[4])
plt.show()
```
